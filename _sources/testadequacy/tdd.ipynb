## TDD

Test-Driven Develoment has become a very popular pratice among developers. But, what is TDD? The concept is surprisingly simple, before writting the implementation code, write the test to prove whether your implementation works or not. The conventional approach makes us write the feature and then the test, with TDD approach the process is the opposite.
